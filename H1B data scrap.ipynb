```python
## Web scrap script

import requests

page = requests.get("https://h1bdata.info/index.php?em=&job=Data+Scientist&city=&year=2018")
page
```

```python
page.status_code
# successful download
```

```python
# parsing the page with beautifulSoup

from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'lxml')
```

```python
import pandas as pd

table = soup.find_all('table')[0]
df = pd.read_html(str(table))[0]
employer = df['EMPLOYER'].tolist()
job_title = df['JOB TITLE'].tolist()
base_sal = df['BASE SALARY'].tolist()
location = df['LOCATION'].tolist()
sub_date = df['SUBMIT DATE'].tolist()
srt_date = df['START DATE'].tolist()
case_status = df['CASE STATUS'].tolist()
```

```python
len(employer)
```

```python
sal_table = pd.DataFrame({'employer':employer,
                         'job_title':job_title,
                         'base_sal':base_sal,
                         'location':location,
                         'sub_date':sub_date,
                         'srt_date':srt_date,
                         'case_status':case_status})
sal_table.sort_values('base_sal', ascending = False, inplace = True)
sal_table
```
